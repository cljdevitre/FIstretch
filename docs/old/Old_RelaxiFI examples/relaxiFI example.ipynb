```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import DiadFit as pf
```

```python
res_ascent_1ms = pf.stretch_in_ascent(R_m=1*10**-6, b_m=0.0001,T_K=1473.15,ascent_rate_ms=1,depth_path_ini_fin_step=[10,0,20],EOS='SW96')

decomp_ascent_1=abs(res_ascent_1ms['Pexternal(MPa)'][1]-res_ascent_1ms['Pexternal(MPa)'][0])/res_ascent_1ms['Time(s)'][1]
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
depth_path_ini_fin_step=[10,0,20]
ascent_rate_ms=1

fig, (ax0,ax1) = plt.subplots(1,2, figsize=(10,3))
ax0.plot(res_ascent_1ms['Depth(km)'],res_ascent_1ms['\u0394R/R0 (fractional change in radius)'],marker='s',label=f"Ascent Rate = {ascent_rate_ms} m/s")
ax0.set_xlim([depth_path_ini_fin_step[0],depth_path_ini_fin_step[1]])
ax0.set_xlabel("Depth")
ax0.set_ylabel('\u0394R/R0 (fractional change in radius)')

ax1.plot(res_ascent_1ms['Depth(km)'],res_ascent_1ms['CO2_dens_gcm3'],marker='s',label=f"Ascent Rate = {ascent_rate_ms} m/s")
ax1.set_xlim([depth_path_ini_fin_step[0],depth_path_ini_fin_step[1]])
ax1.set_xlabel("Depth")
ax1.set_ylabel("CO2_density_gmL")
ax0.legend(loc='best')
ax1.legend(loc='best')
fig.tight_layout()
plt.show()
fig.savefig("fig4diadfit_paper_relaxifi.pdf",format='pdf')
```

```python
fig, ax0 = plt.subplots(figsize=(6, 3))
ax1 = ax0.twinx()  # Create a twin y-axis that shares the same x-axis

ax0.plot(-res_ascent_1ms['Depth(km)'], res_ascent_1ms['\u0394R/R0 (fractional change in radius)'], marker='s', label='\u0394R/R0 (fractional change in radius)')
ax0.set_xlabel("Depth")
ax0.set_ylabel('\u0394R/R0 (fractional change in radius)')

ax1.plot(-res_ascent_1ms['Depth(km)'], res_ascent_1ms['CO2_dens_gcm3'], marker='s', label='CO2_dens_gcm3',color='tab:red')  # Use a different color for ax1
ax1.set_ylabel("CO2_density_gmL")

# Combine the legends into a single legend
lines0, labels0 = ax0.get_legend_handles_labels()
lines1, labels1 = ax1.get_legend_handles_labels()
ax0.legend(lines0 + lines1, labels0 + labels1, loc='best',title=f"Ascent Rate = {1} m/s")

plt.show()
```
